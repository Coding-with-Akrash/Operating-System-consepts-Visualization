```python
"""
============================================================
  Task 1: Exploring and Visualizing the Iris Dataset
  Role  : AI/ML Engineering Internship
  Author: Maarij Ahmed
  Date of submission  : 28 April 2026
============================================================

Objective:
    Load, inspect, and visualize the classic Iris dataset to
    understand data distributions, feature relationships,
    and potential outliers using pandas, matplotlib, and seaborn.

Dataset:
    The Iris dataset contains 150 samples across 3 species of iris
    flowers (Setosa, Versicolor, Virginica), each described by 4
    numeric features measured in centimeters.
"""

# ─────────────────────────────────────────────
# 1. IMPORTS
# ─────────────────────────────────────────────
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")  # Suppress non-critical warnings for clean output


# ─────────────────────────────────────────────
# 2. CONFIGURATION / STYLING
# ─────────────────────────────────────────────

# Seaborn theme for consistent, publication-quality visuals
sns.set_theme(style="whitegrid", palette="Set2", font_scale=1.1)

# Color palette matched to the three Iris species
SPECIES_PALETTE = {
    "setosa":     "#4CAF50",   # Green
    "versicolor": "#2196F3",   # Blue
    "virginica":  "#FF5722",   # Deep Orange
}

# Output figure resolution
DPI = 150


# ─────────────────────────────────────────────
# 3. LOAD DATASET
# ─────────────────────────────────────────────

def load_iris() -> pd.DataFrame:
    """
    Load the Iris dataset via seaborn's built-in loader.

    Returns
    -------
    pd.DataFrame
        Raw Iris dataframe with 150 rows and 5 columns.

    Notes
    -----
    Alternatively, you can load from a local CSV file:
        df = pd.read_csv("iris.csv")
    """
    df = sns.load_dataset("iris")
    print("✅  Dataset loaded successfully via seaborn.\n")
    return df


# ─────────────────────────────────────────────
# 4. BASIC INSPECTION
# ─────────────────────────────────────────────

def inspect_dataset(df: pd.DataFrame) -> None:
    """
    Perform and print basic structural and statistical
    inspection of the dataframe.

    Parameters
    ----------
    df : pd.DataFrame
        The Iris dataframe to inspect.
    """

    separator = "=" * 55

    # --- Shape ---
    print(separator)
    print("📐  DATASET SHAPE")
    print(separator)
    print(f"  Rows    : {df.shape[0]}")
    print(f"  Columns : {df.shape[1]}\n")

    # --- Column names ---
    print(separator)
    print("🏷️   COLUMN NAMES")
    print(separator)
    for i, col in enumerate(df.columns, start=1):
        print(f"  {i}. {col}")
    print()

    # --- First 5 rows ---
    print(separator)
    print("👀  FIRST 5 ROWS (.head())")
    print(separator)
    print(df.head().to_string(index=True))
    print()

    # --- Data types and non-null counts ---
    print(separator)
    print("ℹ️   DATASET INFO (.info())")
    print(separator)
    df.info()
    print()

    # --- Descriptive statistics ---
    print(separator)
    print("📊  DESCRIPTIVE STATISTICS (.describe())")
    print(separator)
    print(df.describe().round(2).to_string())
    print()

    # --- Species distribution ---
    print(separator)
    print("🌸  SPECIES DISTRIBUTION")
    print(separator)
    print(df["species"].value_counts().to_string())
    print()


# ─────────────────────────────────────────────
# 5. SCATTER PLOT — Feature Relationships
# ─────────────────────────────────────────────

def plot_scatter(df: pd.DataFrame) -> None:
    """
    Create a scatter plot matrix (pair plot) to visualize
    pairwise relationships between all numeric features,
    colour-coded by species.

    A pair plot is the fastest way to spot:
        - Linear/non-linear correlations between features.
        - Natural clustering of species groups.

    Parameters
    ----------
    df : pd.DataFrame
        The Iris dataframe.
    """
    print("📈  Generating Scatter Plot (Pair Plot)...")

    fig = sns.pairplot(
        data=df,
        hue="species",            # Color points by species label
        palette=SPECIES_PALETTE,
        diag_kind="kde",          # Kernel Density Estimate on the diagonal
        plot_kws={"alpha": 0.7, "s": 60, "edgecolor": "white", "linewidth": 0.5},
        diag_kws={"fill": True},
    )

    fig.figure.suptitle(
        "Iris Dataset — Pairwise Feature Relationships",
        y=1.02, fontsize=15, fontweight="bold"
    )

    # Annotate purpose below title
    fig.figure.text(
        0.5, 1.0,
        "Each point represents one flower sample; colours indicate species",
        ha="center", fontsize=10, color="gray"
    )

    plt.tight_layout()
    plt.savefig("scatter_pairplot.png", dpi=DPI, bbox_inches="tight")
    print("   ✅  Saved → scatter_pairplot.png\n")
    plt.show()


# ─────────────────────────────────────────────
# 6. HISTOGRAMS — Value Distributions
# ─────────────────────────────────────────────

def plot_histograms(df: pd.DataFrame) -> None:
    """
    Plot overlapping histograms for each numeric feature,
    split by species, to reveal how values are distributed
    and whether species are separable per feature.

    Parameters
    ----------
    df : pd.DataFrame
        The Iris dataframe.
    """
    print("📊  Generating Histograms...")

    # Identify the four numeric feature columns
    numeric_features = df.select_dtypes(include="number").columns.tolist()

    fig, axes = plt.subplots(
        nrows=2, ncols=2,
        figsize=(12, 8),
        facecolor="#FAFAFA"
    )
    fig.suptitle(
        "Iris Dataset — Feature Value Distributions by Species",
        fontsize=14, fontweight="bold", y=1.01
    )

    for ax, feature in zip(axes.flatten(), numeric_features):

        # Plot a separate histogram for each species on the same axis
        for species, color in SPECIES_PALETTE.items():
            subset = df[df["species"] == species][feature]
            ax.hist(
                subset,
                bins=15,
                alpha=0.55,          # Transparency so overlapping bars show through
                color=color,
                edgecolor="white",
                linewidth=0.6,
                label=species.capitalize()
            )

        # Formatting
        ax.set_title(feature.replace("_", " ").title(), fontsize=12, fontweight="semibold")
        ax.set_xlabel("Value (cm)", fontsize=10)
        ax.set_ylabel("Frequency", fontsize=10)
        ax.legend(fontsize=9)
        ax.grid(axis="y", alpha=0.4)

    plt.tight_layout()
    plt.savefig("histograms.png", dpi=DPI, bbox_inches="tight")
    print("   ✅  Saved → histograms.png\n")
    plt.show()


# ─────────────────────────────────────────────
# 7. BOX PLOTS — Outlier Detection
# ─────────────────────────────────────────────

def plot_boxplots(df: pd.DataFrame) -> None:
    """
    Generate box plots for each feature grouped by species to:
        - Display the median, IQR (interquartile range), and whiskers.
        - Highlight individual outlier points beyond 1.5× IQR.
        - Compare spread and central tendency across species.

    Parameters
    ----------
    df : pd.DataFrame
        The Iris dataframe.
    """
    print("📦  Generating Box Plots...")

    numeric_features = df.select_dtypes(include="number").columns.tolist()

    fig, axes = plt.subplots(
        nrows=2, ncols=2,
        figsize=(12, 8),
        facecolor="#FAFAFA"
    )
    fig.suptitle(
        "Iris Dataset — Box Plots: Spread & Outliers by Species",
        fontsize=14, fontweight="bold", y=1.01
    )

    for ax, feature in zip(axes.flatten(), numeric_features):

        sns.boxplot(
            data=df,
            x="species",
            y=feature,
            palette=SPECIES_PALETTE,
            width=0.5,
            linewidth=1.4,
            flierprops=dict(          # Style the outlier dots
                marker="o",
                markerfacecolor="red",
                markeredgecolor="darkred",
                markersize=6,
                alpha=0.7,
            ),
            ax=ax,
        )

        # Overlay individual data points (strip plot) for context
        sns.stripplot(
            data=df,
            x="species",
            y=feature,
            color="black",
            size=3,
            alpha=0.3,
            jitter=True,
            ax=ax,
        )

        # Formatting
        ax.set_title(feature.replace("_", " ").title(), fontsize=12, fontweight="semibold")
        ax.set_xlabel("Species", fontsize=10)
        ax.set_ylabel("Value (cm)", fontsize=10)
        ax.grid(axis="y", alpha=0.4)

    plt.tight_layout()
    plt.savefig("boxplots.png", dpi=DPI, bbox_inches="tight")
    print("   ✅  Saved → boxplots.png\n")
    plt.show()


# ─────────────────────────────────────────────
# 8. BONUS — Correlation Heatmap
# ─────────────────────────────────────────────

def plot_correlation_heatmap(df: pd.DataFrame) -> None:
    """
    Visualize the Pearson correlation matrix as a heatmap.

    Correlation values range from -1 (perfect negative) to +1
    (perfect positive). This helps identify which feature pairs
    move together — useful context before building ML models.

    Parameters
    ----------
    df : pd.DataFrame
        The Iris dataframe.
    """
    print("🌡️   Generating Correlation Heatmap...")

    # Compute correlation matrix on numeric columns only
    corr_matrix = df.select_dtypes(include="number").corr()

    fig, ax = plt.subplots(figsize=(7, 5), facecolor="#FAFAFA")

    sns.heatmap(
        corr_matrix,
        annot=True,              # Show numeric correlation values in each cell
        fmt=".2f",               # 2 decimal places
        cmap="coolwarm",         # Diverging color: red = positive, blue = negative
        linewidths=0.5,
        linecolor="white",
        vmin=-1, vmax=1,         # Fix the color scale to the full range
        square=True,
        ax=ax,
    )

    ax.set_title(
        "Feature Correlation Matrix (Pearson)",
        fontsize=13, fontweight="bold", pad=14
    )

    plt.tight_layout()
    plt.savefig("correlation_heatmap.png", dpi=DPI, bbox_inches="tight")
    print("   ✅  Saved → correlation_heatmap.png\n")
    plt.show()


# ─────────────────────────────────────────────
# 9. MAIN PIPELINE
# ─────────────────────────────────────────────

def main() -> None:
    """
    Orchestrates the full data exploration pipeline:
        1. Load dataset
        2. Inspect structure and statistics
        3. Scatter plot  → feature relationships
        4. Histograms    → value distributions
        5. Box plots     → outlier detection
        6. Heatmap       → feature correlations (bonus)
    """

    print("\n" + "=" * 55)
    print("   IRIS DATASET — EXPLORATION & VISUALIZATION")
    print("=" * 55 + "\n")

    # Step 1 — Load
    df = load_iris()

    # Step 2 — Inspect
    inspect_dataset(df)

    # Step 3 — Scatter plot
    plot_scatter(df)

    # Step 4 — Histograms
    plot_histograms(df)

    # Step 5 — Box plots
    plot_boxplots(df)

    # Step 6 — Correlation heatmap (bonus insight)
    plot_correlation_heatmap(df)

    print("=" * 55)
    print("   ✅  All tasks completed successfully!")
    print("   📁  Check the current directory for saved PNGs.")
    print("=" * 55 + "\n")


# ─────────────────────────────────────────────
# ENTRY POINT
# ─────────────────────────────────────────────

if __name__ == "__main__":
    main()
```
